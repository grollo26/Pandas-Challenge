#### 3.2 The school wants to change scoring system, and replace 1-10 with a letter grade. Change 10 to A, 9/8 to B, 7/6 to C and any lower scores to D.

```python
def convert(score): # this converts numeric scores into letters
    if score== 10:
        return 'A'
    elif score >=8:
        return 'B'
    elif score>=6:
        return 'C'
    else:
        return 'D'
```

```python
convert(7) # it works on 7 :)
```

```python
subjects=['L1','L2','Maths','Philosophy']
```

```python
for x in subjects:
    df2[x]=df2[x].apply(convert)
```

```python
df2.head()
```
